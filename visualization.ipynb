```python
import pandas as pd
import os
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
plt.style.use('seaborn')
%matplotlib inline
```

```python
df_prov = []
direct = 'C:/Users/Bruno Aquino/Documents/Kaggle Quora/data/train/'
for file in [x for x in os.listdir(direct)]:
    df_prov.append(pd.read_csv(direct + file))
df = pd.concat(df_prov)
del df_prov
```

```python
df['chunk_sim'] = np.clip(df['chunk_sim'], 0, 1)
```

```python
df_0 = df[df['is_duplicate'] == 0]
df_1 = df[df['is_duplicate'] == 1]
```

```python
print((len(df_0), len(df_1)))
```

```python
sub = 331
for column in ['match_words', 'match_sub_root', 'match_set', 'n_words', 'chunk_sim', 'n_verb', 'word_importance', 'match_ngram']:
    sp = plt.subplot(sub)
    a0 = plt.hist(df_0[column], color='r', alpha=0.5)
    a1 = plt.hist(df_1[column], color='b', alpha=0.5)
    t = plt.title(column)
    sub += 1
plt.subplots_adjust(top=2, bottom=0.08, left=0, right=2, hspace=0.2,
                    wspace=0.2)
```
